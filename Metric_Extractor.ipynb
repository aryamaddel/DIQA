## BRISQUE

```python
from brisque import BRISQUE
import numpy as np
from PIL import Image

img_path = r"C:\Users\user\Documents\GitHub\NR-IQA\koniq10k_1024x768\1024x768\9895608016.jpg"
print(f"Image path: {img_path}")
img = Image.open(img_path)
ndarray = np.asarray(img)
print(f"Image ndarray shape: {ndarray.shape}")

obj = BRISQUE(url=False)
score = obj.score(img=ndarray)
print(f"BRISQUE score: {score}")
```

##  PIQE

```python
from pypiqe import piqe

score, activityMask, noticeableArtifactMask, noiseMask = piqe(ndarray)

print(f"PIQE score: {score}")
print(f"Activity Mask: {activityMask}")
print(f"Noticeable Artifact Mask: {noticeableArtifactMask}")
print(f"Noise Mask: {noiseMask}")
```
